```python
import math
from matplotlib import pyplot as plt
import numpy as np
```

# **Zadanie 1.**

a) grupa 1:

```python
def f1(x):
    return (x**0.999999)*math.log(x)

def f2(x):
    return 10000000*x

def f3(x):
    return 1.000001**x

def f4(x):
    return x**2

x = np.arange(100000.,100000000.,100)

plt.plot(x, [f1(i) for i in x], "r", label="f1")
plt.plot(x, [f2(i) for i in x], "b", label="f2")
plt.plot(x, [f3(i) for i in x], "g", label="f3")
plt.plot(x, [f4(i) for i in x], "y", label="f4")

plt.legend()
plt.grid()
plt.show()
```

b) grupa 2:

```python
def f1(x):
    return 2**(100*x)

def f2(x):
    return math.factorial(x)/(2*math.factorial(x-2))

def f3(x):
    return x*math.sqrt(x)

x = range(2,10)

plt.plot(x, [f1(i) for i in x], "r", label="f1")
plt.plot(x, [f2(i) for i in x], "b", label="f2")
plt.plot(x, [f3(i) for i in x], "g", label="f3")

plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x, [f2(i) for i in x], "b", label="f2")
plt.plot(x, [f3(i) for i in x], "g", label="f3")

plt.legend()
plt.grid()
plt.show()
```

c) grupa 3:

```python
def f1(x):
    return x**(math.sqrt(x))

def f2(x):
    return 2**x

def f3(x):
    return x**10*2**(x/2)

def f4(x):
    return sum(i+1 for i in range(x+1))

x = range(1,10)

plt.plot(x, [f1(i) for i in x], "r", label="f1")
plt.plot(x, [f2(i) for i in x], "b", label="f2")
plt.plot(x, [f3(i) for i in x], "g", label="f3")
plt.plot(x, [f4(i) for i in x], "y", label="f4")

plt.legend()
plt.grid()
plt.show()
```
